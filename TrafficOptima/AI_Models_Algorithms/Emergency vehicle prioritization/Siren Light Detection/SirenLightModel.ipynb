```python
import torch
import cv2
import numpy as np
import pandas as pd
# Step 2: Load the YOLOv8 model
model = YOLO("best.pt")

# Step 3: Capture and preprocess frames
cap = cv2.VideoCapture('videoplayback4.mp4')  # Replace 'video.mp4' with your video file path

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess frame (you can adjust the preprocessing steps as needed)
    resized_frame = cv2.resize(frame, (640, 480))  # Resize frame
    normalized_frame = resized_frame.astype(np.float32) / 255.0  # Normalize pixel values

    # Step 4: Detect sirens using YOLOv8
    results = model.predict(normalized_frame)
    print("ress",results[0])
    
    # Extract bounding boxes, class labels, and confidences
    bboxes = results.pandas().xyxy[0][['xmin', 'ymin', 'xmax', 'ymax']].values
    confidences = results.pandas().xyxy[0]['confidence'].values
    class_labels = results.pandas().xyxy[0]['name'].values


    for bbox, confidence, class_label in zip(bboxes, confidences, class_labels):
        if class_label == 0:  # Assuming class 0 represents the siren
            # Step 5: Classify the siren light state
            x1, y1, x2, y2 = bbox.astype(np.int)
            roi = frame[y1:y2, x1:x2]  # Extract the region of interest (ROI)

            # Apply image processing techniques for state classification
            # Example: Apply color-based segmentation to classify the siren light state
            hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
            lower_red = np.array([0, 50, 50])
            upper_red = np.array([10, 255, 255])
            mask = cv2.inRange(hsv, lower_red, upper_red)
            red_pixels = cv2.countNonZero(mask)
            light_state = "On" if red_pixels > 0 else "Off"

            # Step 6: Display or process the results
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f"Light State: {light_state}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow("Siren Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
